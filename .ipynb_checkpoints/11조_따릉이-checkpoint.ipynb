# 11조_따릉이의 이용시간에 따른 최저기준시간 변경

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import font_manager
fontprop = font_manager.FontProperties(fname="c:\Windows\Fonts\malgun.ttf").get_name()
%matplotlib inline
```

```python
data = pd.read_excel('서울특별시 공공자전거 시간대별 대여정보_201811.xlsx', header = 0)
data
```

```python
One_hour = data[(data['이동시간']<=60) & (data['대여구분코드'] != '정기권')] # 사용시간이 1시간 밑인 데이터만 추출
One_hour = One_hour.reset_index(drop=True)
One_hour
```

```python
One_hour.shape
```

```python
time = One_hour['대여시간'].value_counts() # 대여시각에 따른 이용자의 수 count
```

```python
plt.grid()
plt.scatter(x = time.index, y = time.values) # x축이 대여시각,  y축이 이용자의 수(명)
plt.xlabel('대여시간(시각)', fontproperties = fontprop)
plt.ylabel('이용자의 수(명)', fontproperties = fontprop)
plt.title('대여시각에 따른 이용자의 수', fontproperties = fontprop)
```

```python
time_5 = One_hour[(One_hour['대여시간']==15) | (One_hour['대여시간']==16) | (One_hour['대여시간']==17) | (One_hour['대여시간']==18) | (One_hour['대여시간']==19)]
time_5 # 가장 높은 이용자의 수가 대여한 시간 상위 5개(15~19)
```

```python
user = time_5['이동시간'].value_counts() # 이용시간에 따른 이용자의 수 count
```

```python
plt.grid()
plt.scatter(x = user.index, y = user.values) # x축이 이용시간(분),  y축이 이용시간 count값
plt.xlabel('이용시간(분)', fontproperties = fontprop)
plt.ylabel('이용자의 수(명)', fontproperties = fontprop)
plt.title('이용시간에 따른 이용자의 수', fontproperties = fontprop)
```

```python
sum(user.index * user.values / len(time_5)) # 평균이 24분
```

평균으로 나온 값을 그대로 사용하기는 심리적으로 불편한 숫자라고 생각하고 24분이 아니라 딱 떨어지는 숫자인 30분으로 정하면 어떨까? 생각해봄.

```python
user.values[user.index <= 30].sum() / len(time_5) # 30분을 기준으로 했을 때 그 이하가 약 70%
```

결론 : 가장 많이 이용하는 시간대에 30분 이용권을 만들어서 가격적으로 이득을 보는 사람들이 많아졌으면 좋겠다.
